```python
# Pandas importing. 
# import pandas as pd
# Importing matplotlib
import matplotlib.pyplot as plt


# Creating data frame. 
df = pd.read_csv('grocery.csv')


# Create a line plot of four measures. 
df.plot()
```

```python
# Create a scatter plot of two measures. 

df.plot.scatter(x="petal_length", y="petal_width")

# Showing the scatter plot
plt.show()
```

```python
# Creating a box plot of two measures. 
df.plot.box()

# Showing the box plot
plt.show()
```

```python
# Creating the plot. 
df.plot.bar()

# Showing the plot. 
plt.show()
```

```python
# Create the plot. 
df.plot.bar(subplots=True)

# Showing the plot. 
plt.show()
```

```python
# Creating the plot. 
df.plot.hist(subplots=True)

# Showing the plot. 
plt.show()
```
