```python
#import libraries
import pandas as pd
import plotly as pltly
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
 
# Load the iris dataset
iris = sns.load_dataset('iris')
 
# Plot the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=iris, x='sepal_length', y=iris.index, hue='species', markers=True, dashes=False)
 
# Add title and labels
plt.title('Sepal Length Across Iris Species')
plt.xlabel('Sepal Length')
plt.ylabel('Sample Index')
 
# Display the plot
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
 
# Load the iris dataset
iris = sns.load_dataset('iris')
 
# Define custom colors for each species
species_colors = {'setosa': 'blue', 'versicolor': 'orange', 'virginica': 'green'}
 
# Create a scatterplot with species colored
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette=species_colors, alpha=0.7)
 
# Add labels and title
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Scatterplot of Sepal Length vs. Sepal Width with Species Colored')
 
# Show the plot
plt.legend(title='Species')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
# Load iris dataset from seaborn
iris = sns.load_dataset('iris')
 
# Count measurements by species
species_counts = iris['species'].value_counts()
 
# Plotting the pie chart
colors = {'setosa': 'blue', 'versicolor': 'red', 'virginica': 'green'}
plt.figure(figsize=(8, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140, colors=[colors[species] for species in species_counts.index])
plt.title('Measurement Counts by Species')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
 
# Load iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
 
# Define bins for sepal length
bins = [4.3, 5.2, np.inf]
nam = ['Sepal-Length 4.3-5.2', 'More than 5.2']
df['sepal_length_bins'] = pd.cut(df['sepal length (cm)'], bins, labels=nam, include_lowest=True)
 
# Count occurrences in each bin
counts = df['sepal_length_bins'].value_counts()
 
# Set text color
plt.rcParams['text.color'] = 'k'
 
# Colors and styles
colors = ['lightblue', 'lightcoral']
explode = (0.1, 0)
 
# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode)
plt.title('Distribution of Sepal Lengths', fontsize=16)
 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
 
# Load iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
 
# Define bins for sepal length
bins = [4.3, 5.2, np.inf]
nam = ['Sepal-Length 4.3-5.2','More than 5.2']
df['Range'] = pd.cut(df['sepal length (cm)'], bins, labels=nam, include_lowest=True)
 
# Set text color
plt.rcParams['text.color'] = 'k'
 
names=['Iris-Setosa', 'Iris-Versicolor', 'Iris-Virginica']
size = df['species'].value_counts().tolist()
size_1 = df.groupby(['species', 'Range']).size().tolist()
 
names_1=['S1', 'S2', 'Ver1', 'Ver2', 'Vir1', 'Vir2']
 
# Colors
a, b, c = [plt.cm.Blues, plt.cm.Reds, plt.cm.Greens]
 
# Outside Ring
fig, ax = plt.subplots(figsize=(6,6))
ax.axis('equal')
mypie, _ = ax.pie(size, radius=1.3, labels=names, colors=[a(0.6), b(0.6), c(0.6)], textprops={'fontsize': 16})
plt.setp(mypie, width=0.3, edgecolor='white')
 
# Inside Ring
mypie2, _ = ax.pie(size_1, radius=1.3 - 0.3, labels=names_1, labeldistance=0.8,
                   colors=[a(0.5), a(0.4), b(0.5), b(0.4), c(0.3), c(0.2)],
                   textprops={'fontsize': 16})
plt.setp(mypie2, width=0.4, edgecolor='white')
plt.margins(5, 5)
plt.title("Sepal Length in different Species", fontsize=16, pad=20)
plt.legend(mypie2, ('Length 4.3-5.2 in Setosa', 'more than 5.2 in Setosa',
                    'Length 4.3-5.2 in Versicolor', 'more than 5.2 in Versicolor',
                    'Length 4.3-5.2 in Virginica', 'more than 5.2 in Virginica'),
           bbox_to_anchor=(0, 1), loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd
 
# Load iris dataset
iris = load_iris()
 
# Convert to DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
 
# Create boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal length (cm)', data=df)
plt.title('Boxplot of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```
