```python
# Ramon Montoya
# Assignment 3B: Creating Data Visualizations 
# Parts 1 and 2

# Importing the libraries 
import pandas as pd
import plotly as pltly
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the Titanic dataset as a dataframe titanic_data.
titanic_data = pd.read_csv("titanic_train.csv")

# Summarize the dataset. 
print(titanic_data.info())

# Use the following code to plot a histogram of ages of Titanic Passengers
# Plot the histogram of ages
plt.figure(figsize=(10, 6))
plt.hist(titanic_data['Age'].dropna(), color='green', edgecolor='black')
plt.title('Histogram of Ages of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Use the following code to create a bar chart of Passengers with Parents or Children Onboard.
# Create a bar chart of Passengers with Parents or Children Onboard
# Count the frequency of each unique value in the 'Parch' column
parch_counts = titanic_data['Parch'].value_counts()
 
# Plot bar chart of passengers with parents or children onboard
plt.figure(figsize=(10, 6))
plt.bar(parch_counts.index, parch_counts, color='yellow', edgecolor='black')
plt.title('Passengers with Parents or Children Onboard')
plt.xlabel('Number of Parents or Children')
plt.ylabel('Frequency')
plt.show()

```
